```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.compose import make_column_transformer
```

```python
import pandas as pd

df = pd.DataFrame({'Salary':[25000,48000,71000,85000,90000,55000],'City':['Bangalore','Delhi','Hyderabad','Bangalore','Hyderabad','Bangalore']
                  ,'Gender':['Male','Female','Female','Female','Male','Male'],'Exp':[1,3,5,6,9,None]})
df
```

```python
ohe = OneHotEncoder()
si = SimpleImputer()
```

```python
ct = make_column_transformer((ohe,['City','Gender']),(si,['Exp']),remainder='passthrough')
```

```python
encoded = pd.DataFrame(ct.fit_transform(df))
```

```python
encoded
```

```python
encoded = pd.DataFrame(ct.fit_transform(df),columns=['City_Bangalore','City_Delhi','City_Hyderabad','Gender_male','Gender_Female','Exp','Salary'])
encoded
```

```python
from sklearn.preprocessing import OrdinalEncoder
import pandas as pd
employee = pd.DataFrame({'position':['SE','Manager','Team Lead','SSE'],
                        'project':['A','B','C','D'],
                        'Salary':[25000,85000,71000,48000]})
employee
```

```python
ord_enc = OrdinalEncoder(categories=[['SE','SSE','Team Lead','Manager'],['A','B','C','D']])
encoded_df = ord_enc.fit_transform(employee[['position','project']])
encoded_df
```

```python
import pandas as pd

df = pd.DataFrame({'Salary':[25000,48000,71000,85000,90000,49000],'City':['Bangalore','Delhi','Hyderabad','Bangalore','Hyderabad','Bangalore']
                  ,'Gender':['Male','Female','Female','Female','Male','Male'],'Exp':[1,3,5,6,9,None]})
df
```

```python
from sklearn.impute import KNNImputer
```

```python
knn = KNNImputer(n_neighbors=3)
knn_imp = pd.DataFrame(knn.fit_transform(df[['Salary','Exp']]))
knn_imp
```

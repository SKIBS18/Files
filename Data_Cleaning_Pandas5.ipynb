```python
raw_data = {'First_Name':['Jason',np.nan,'Tina','Jake','Ammy','Madhu'],
           'Last_Name':['Miller',np.nan,'Ali','Milner','Cooze','Patil'],
           'Age':[42,np.nan,36,24,73,25],'Sex':['M',np.nan,'F','M','F','F'],
           'Unit_1':[4,np.nan,np.nan,2,3,4],'Unit_2':[3,np.nan,np.nan,4,4,4]}
df =pd.DataFrame(raw_data,columns = ['First_Name','Last_Name','Age','Sex','Unit_1','Unit_2'])
df
```

```python
drop_missing = df.dropna()
drop_missing
```

```python
df
```

```python
#Rows having all Nan
data_cleaned = df.dropna(how='all')
data_cleaned
```

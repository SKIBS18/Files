```python
#DateTimeManipulation
import numpy as np
import pandas as pd
```

```python
date_range = pd.date_range('4/12/2021',periods = 12,freq ='D')
dates = pd.DataFrame(date_range,columns=['Date'])
```

```python
dates
```

```python
date_range = pd.date_range('4/12/2021',periods = 12,freq ='M')
dates = pd.DataFrame(date_range,columns=['Date'])
dates
```

```python
date_range = pd.date_range('4/12/2021',periods = 12,freq ='Y')
dates = pd.DataFrame(date_range,columns=['Date'])
dates
```

```python
dates['Day'] = dates['Date'].dt.day
dates['Month'] = dates['Date'].dt.month
dates['Year'] = dates['Date'].dt.year

dates
```

```python
del dates['Date']
dates
```

```python
df = pd.DataFrame({'Date_Time':['Jan 1 12:00 am','Jan 2 12:00 am']})
df
```

```python
df['Date'] = pd.to_datetime(df['Date_Time'],format = '%b %d %I:%M %p')
df
```

```python
df['Day'] = df['Date'].dt.day
df['Month'] = df['Date'].dt.month
df['Year'] = df['Date'].dt.year

df
```

```python
del df['Date_Time']
df
```
